**1) ¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?** 

```python
#Creamo un array de 1000 elmentos booleanos usando linspace
array1000 = np.linspace(True, True, 1000, dtype = bool)

#Obtenemos el tamaño en bytes del array
arrayTotal = array1000.nbytes

print("Tamaño en bytes de array de 1000 elementos booleanos es: ", arrayTotal, "bytes" )
print(array1000)


```

**4) Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], 
obtenga la media aritmética de la diferencia (a-b).** 

```python
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

diferencia = a - b
media = np.mean(diferencia)

print("La media aritmética de la diferencia es:", media)
```

**6) Copie el array del ejercicio anterior. Traspóngalo (. 
Agregue ahora dos nuevos pares de coordenadas: (10.8, 
3.0) y (35.8, 12.0).**

```python
array_original = np.array([[1.33, 4.5],
                            [30.0, 10.7],
                            [70.2, 0.5]])

array_traspuesto = array_original.T

nuevas_coordenadas = np.array([[10.8, 3.0],
                                [35.8, 12.0]])

array_traspuesto = np.concatenate((array_traspuesto, nuevas_coordenadas.T), axis=1)

nuevo_array = array_traspuesto.T

print("Dimensiones del nuevo array:", nuevo_array.shape)
```

**2) Realice una gráfica de barras sobre la polbación de cada 
ciudad, ordenela de menor a mayo y responda:** 
 
**a. ¿Cuál es la segunda ciudad más poblada? 
b. ¿Qué posición ocupa Granada en el ranking de

```python
## Se crea una tabla para saber todas las pronvincias 

df_ordenado = dataset.sort_values(by='Poblacion')
provincias_unicas = dataset['Provincia'].unique()
df_provincias = pd.DataFrame(provincias_unicas, columns=['Provincia'])
print(df_provincias)
```

```python
##Debido a la gran cantidad de de ciudades 852 se creo 51 graficos de las provincias que contienen 
## la poblacion de cada uno de las ciudades de mayor poblacion a menor poblacion


provincias_unicas = dataset['Provincia'].unique()

for provincia in provincias_unicas:
    # Filtrar el DataFrame por la provincia actual
    df_provincia = dataset[dataset['Provincia'] == provincia]
    

    df_provincia = df_provincia.sort_values(by='Poblacion', ascending=False)
    plt.figure(figsize=(10, 6))
    plt.barh(df_provincia['Municipio'], df_provincia['Poblacion'])
    plt.xlabel('Población')
    plt.title(f'Ciudades de {provincia} por población')
    plt.gca().invert_yaxis()  # Invertir el eje y para mostrar de mayor a menor población
    plt.tight_layout()
```

```python
df_sorted = dataset.sort_values(by='Poblacion', ascending=False)


segundo_municipio_mas_poblado = df_sorted.iloc[1]['Municipio']
poblacion_segundo_municipio = df_sorted.iloc[1]['Poblacion']

print(f"El segundo municipio más poblado es {segundo_municipio_mas_poblado} con una población de {poblacion_segundo_municipio} habitantes.")
```

```python
df_sorted = dataset.sort_values(by='Poblacion', ascending=False)

top_20_municipios = df_sorted.head(20)


plt.figure(figsize=(12, 8))
plt.barh(top_20_municipios['Municipio'], top_20_municipios['Poblacion'])
plt.xlabel('Población')
plt.title('Top 20 Municipios más poblados')
plt.gca().invert_yaxis()  # Invertir el eje y para mostrar de mayor a menor población
plt.tight_layout()


plt.show()
```

```python
municipio_norte = dataset.loc[dataset['Y'].idxmax()]

provincia_municipio_norte = municipio_norte['Provincia']
poblacion_municipio_norte = municipio_norte['Poblacion']


print(f"El municipio más al Norte es {municipio_norte['Municipio']} en la provincia de {provincia_municipio_norte} con una población de {poblacion_municipio_norte} habitantes.")
```

**5) Obtenga la media, mediana, desviación estándar, valor 
máximo y valor mínimo de la población de los municipios 
de la provincia de Granada.**

```python
municipios_granada = dataset[dataset['Provincia'] == 'Granada']


media_poblacion = municipios_granada['Poblacion'].mean()
mediana_poblacion = municipios_granada['Poblacion'].median()
desviacion_estandar_poblacion = municipios_granada['Poblacion'].std()
maximo_poblacion = municipios_granada['Poblacion'].max()
minimo_poblacion = municipios_granada['Poblacion'].min()


print(f"Media de población: {media_poblacion}")
print(f"Mediana de población: {mediana_poblacion}")
print(f"Desviación estándar de población: {desviacion_estandar_poblacion}")
print(f"Valor máximo de población: {maximo_poblacion}")
print(f"Valor mínimo de población: {minimo_poblacion}")
```

**6) Realice un histograma con la población de los 
Municipios para cada una de las provincias.** 

```python
poblacion_por_provincia = dataset.groupby('Provincia')['Poblacion'].sum()

poblacion_por_provincia = poblacion_por_provincia.sort_values(ascending=False)


plt.figure(figsize=(10, 6))
sns.barplot(x=poblacion_por_provincia.index, y=poblacion_por_provincia.values, palette='viridis')
plt.xlabel('Provincia')
plt.ylabel('Población')
plt.title('Población de los Municipios por Provincia (Ordenado de Mayor a Menor)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()


plt.show()
```

**7) Seleccione al azar cincuenta municipios diferentes de 
entre los diponibles en el archivo. Asegúrese de que no 
se repitan. ¿Luego calcule el promedio de la población y 
la desviación estándar de esto 50 municipios?.**

```python
municipios_aleatorios = dataset.sample(n=51, replace=False, random_state=42)


promedio_poblacion = np.mean(municipios_aleatorios['Poblacion'])
desviacion_estandar_poblacion = np.std(municipios_aleatorios['Poblacion'])


print("Promedio de la población de los 50 municipios seleccionados al azar:", promedio_poblacion)
print("Desviación estándar de la población de los 50 municipios seleccionados al azar:", desviacion_estandar_poblacion)
```

8) Determine los nombres de los Municipios más cercano y 
más lejano a Madrid. Para ello debe calcular la 
distancia en todos ellos y Madrid. No considere a Madrid 
en el análisis ya que la distancia sería cero.

```python
madrid_data = dataset[dataset["Municipio"] == "Madrid"]
madrid_coordsX = (madrid_data["X"].values[0])
madrid_coordsY = (madrid_data["Y"].values[0])

madrid_coords = (madrid_coordsY, madrid_coordsX)

dataset["Distancia_a_Madrid"] = ((dataset["X"] - madrid_coords[0])**2 + (dataset["Y"] - madrid_coords[1])**2)**0.5


municipio_mas_cercano = dataset.loc[dataset["Distancia_a_Madrid"].idxmin()]["Municipio"]
municipio_mas_lejano = dataset.loc[dataset["Distancia_a_Madrid"].idxmax()]["Municipio"]

print("Municipio más cercano a Madrid:", municipio_mas_cercano)
print("Municipio más lejano a Madrid:", municipio_mas_lejano)

```
